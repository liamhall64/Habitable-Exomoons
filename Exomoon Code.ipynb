## Filter Archive

```python
# IMPORTS AND PREAMBLE
import pandas as pd
```

```python
# READ IN CSV FILE
archive = pd.read_csv(r'C:\ ') # YOUR PATH TO CSV FILE


# LISTING ALL COLUMNS FOR EASY ACCESS TO COLUMNS
list(archive.columns.values.tolist()) 
#print('Done.')
```

```python
# FILTERING OUT RADII LESS THAN 0.5
archive.radj = archive['pl_radj']
rad_filt = archive[archive.radj >= 0.5] # KEEPS ROWS THAT HAVE RADIUS GREATER THAN OR EQUAL TO 0.5 JUPITER RADII
#rad_filt
print('Done.')
```

```python
#pd.set_option('display.max_columns', None) # PRINTS WHOLE TABLE OF COLUMNS
#pd.set_option('display.max_rows', None) # PRINTS WHOLE TABLE OF ROWS

rad_filt.method = rad_filt['discoverymethod'] # GIVES 'discoverymethod' COLUMN OWN DATAFRAME
disc_and_rad = rad_filt[rad_filt['discoverymethod'].str.contains('Transit')] # FILTERS ROWS THAT HAVE THE WORD 'transit' IN
disc_and_rad
#print('Done.')
```

## Calculating the CHZ
#### Using Equations from page 10 in Kopparapu et al (2013)
###### https://iopscience.iop.org/article/10.1088/0004-637X/765/2/131/pdf#page=10.44

```python
T = disc_and_rad['st_teff'] - 5780

# Recent Venus
a_v = 1.4316e-4
b_v = 2.9875e-9
c_v = -7.5702e-12
d_v = -1.1635e-15

VS_eff = 1.7753 + (a_v*T) + (b_v*T**2) + (c_v*T**3) + (d_v*T**4)


# Early Mars
a_m = 5.4513e-5
b_m = 1.5313e-9
c_m = -2.7786e-12
d_m = -4.8997e-16

MS_eff = 0.3179 + (a_m*T) + (b_m*T**2) + (c_m*T**3) + (d_m*T**4)

#print(VS_eff, MS_eff)
print('Done.')
```

```python
L = 10**(disc_and_rad['st_lum'])

in_CHZ = (L/VS_eff)**0.5
out_CHZ = (L/MS_eff)**0.5

in_CHZ = pd.DataFrame({'in_CHZ': in_CHZ}) # GIVES TWO NEW COLUMNS OWN HEADER 
out_CHZ = pd.DataFrame({'out_CHZ': out_CHZ}) # GIVES TWO NEW COLUMNS OWN HEADER 
print('Done.')
```

```python
# ADDING THE NEW CHZ COLUMNS TO THE FILTERED DATAFRAME
filtered = pd.concat([disc_and_rad, in_CHZ, out_CHZ], axis=1) # CONCAT IS CONCATENATION - JOINING TWO STRINGS INTO ONE 
filtered
```

Assuming that the orbits follow a Keplarian orbit, the orbit semi-major axis is mean distance from its parent star in AU.

```python
# FILTERS OUT PLANETS WITH ORBITAL SEPARATION OUTSIDE THIS CALCULATED RANGE OF CHZ

filtered_archive = filtered[(filtered['pl_orbsmax'] >= filtered['in_CHZ']) & (filtered['pl_orbsmax'] <= filtered['out_CHZ'])]
filtered_archive
#print('Done.')
```
