## Calculating the CHZ
#### Using Equations from page 10 in Kopparapu et al (2013)
###### https://iopscience.iop.org/article/10.1088/0004-637X/765/2/131/pdf#page=10.44

```python
T = disc_and_mass['st_teff'] - 5780

# Recent Venus
a_v = 1.4316e-4
b_v = 2.9875e-9
c_v = -7.5702e-12
d_v = -1.1635e-15

VS_eff = 1.7753 + (a_v*T) + (b_v*T**2) + (c_v*T**3) + (d_v*T**4)


# Early Mars
a_m = 5.4513e-5
b_m = 1.5313e-9
c_m = -2.7786e-12
d_m = -4.8997e-16

MS_eff = 0.3179 + (a_m*T) + (b_m*T**2) + (c_m*T**3) + (d_m*T**4)

#print(VS_eff, MS_eff)
print('Done.')
```

```python
L = 10**(disc_and_mass['st_lum'])

in_CHZ = (L/VS_eff)**0.5
out_CHZ = (L/MS_eff)**0.5

#print('Inner CHZ:\n{} \n\nOuter CHZ: \n{}'.format(in_CHZ,out_CHZ))
#in_CHZ_df = pd.DataFrame[in_CHZ]
#out_CHZ_df = pd.DataFrame[out_CHZ]
print('Done.')
```

```python
# ADDING THE NEW CHZ COLUMNS TO THE FILTERED DATAFRAME
filtered_df['in_CHZ'] = in_CHZ
filtered_df['out_CHZ'] = out_CHZ

filtered_df
```
